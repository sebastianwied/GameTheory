```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 8
side = 2**n
total = 2**(2*n)
possibilities = np.arange(total)

xcoords = []
ycoords = []
for pos in possibilities:
    x = np.zeros(8, dtype=int)
    y = np.zeros(8, dtype=int)
    for i in range(0,n):
        x[-(i+1)] = 0 if (pos & (2**(2*i))) == 0 else 1
        y[-(i+1)] = 0 if (pos&(2**(2*i+1))) == 0 else 1
    xcoords.append(np.packbits(x, bitorder='big')[0])
    ycoords.append(np.packbits(y, bitorder='big')[0])

plt.figure(figsize=(24,24))
plt.plot(xcoords, ycoords, lw=0.5)
plt.xlim(-1, side)
plt.ylim(side, -1)
```

```python
topRule = [0.7, 0.6, 0.9, 0]
myRule = [0.4, 0.3, 0, .9]
ruleDiff = [(topRule[i] - myRule[i]) for i in range(len(topRule))]
ruleDiffSq = [np.sign(ruleDiff[i])*(ruleDiff[i])**2 for i in range(len(topRule))]
maxChangeId = np.argmax(np.abs(ruleDiff))
ruleDelta = [0 if i != maxChangeId else ruleDiffSq[i] for i in range(len(topRule))]
print(ruleDelta)
```
